```python
#Fixing dates to date time format

# Reload dataframes to ensure original date string formats for conversion
daily_activity = pd.read_csv('/content/dailyActivity_merged.csv')
sleep_minutes = pd.read_csv('/content/minuteSleep_merged.csv')
weight_log_info = pd.read_csv('/content/weightLogInfo_merged.csv')

daily_activity['ActivityDate'] = pd.to_datetime(daily_activity['ActivityDate'], format='%m/%d/%Y')
sleep_minutes['date'] = pd.to_datetime(sleep_minutes['date'], format='%m/%d/%Y %I:%M:%S %p')
weight_log_info['Date'] = pd.to_datetime(weight_log_info['Date'], format='%m/%d/%Y %I:%M:%S %p')

#organising dates from lowest to highest
print("Daily Activity date range:", daily_activity['ActivityDate'].min(), "to", daily_activity['ActivityDate'].max())
print("Sleep Minutes date range:", sleep_minutes['date'].min(), "to", sleep_minutes['date'].max())
print("Weight Log Info date range:", weight_log_info['Date'].min(), "to", weight_log_info['Date'].max())
```

```python
#changing the format of the dates to dd-mm-yyyy
daily_activity['ActivityDate'] = daily_activity['ActivityDate'].dt.strftime('%d-%m-%Y')
sleep_minutes['date'] = sleep_minutes['date'].dt.strftime('%d-%m-%Y')
weight_log_info['Date'] = weight_log_info['Date'].dt.strftime('%d-%m-%Y')
```

### Process Stage

```python
#standardise column names
daily_activity.columns = daily_activity.columns.str.lower().str.replace(' ', '_')
sleep_minutes.columns = sleep_minutes.columns.str.lower().str.replace(' ', '_')
weight_log_info.columns = weight_log_info.columns.str.lower().str.replace(' ', '_')
#verify
print(daily_activity.columns.tolist())
print(sleep_minutes.columns.tolist())
print(weight_log_info.columns.tolist())
```

```python
daily_activity['activitydate']= pd.to_datetime(daily_activity["activitydate"], format='%d-%m-%Y')
daily_activity['day_of_week'] = daily_activity['activitydate'].dt.day_name()
daily_activity['month'] = daily_activity['activitydate'].dt.month_name()

print(daily_activity.dtypes)
```

```python
#weight log
weight_log_info['date']= pd.to_datetime(weight_log_info['date'], format='%d-%m-%Y')
weight_log_info['ismanualreport'] = weight_log_info['ismanualreport'].astype(bool)
print(weight_log_info.dtypes)
```

```python
sleep_minutes['date']=pd.to_datetime(sleep_minutes['date'], format='%d-%m-%Y')
sleep_minutes['date_only'] = sleep_minutes['date'].dt.date

print(sleep_minutes.dtypes)
```

## Handle Null values

```python
print("Nulls in Daily Activity:")
print(daily_activity.isnull().sum())

print("\nNulls in Sleep Minutes:")
print(sleep_minutes.isnull().sum())

print("\nNulls in Weight Log Info:")
print(weight_log_info.isnull().sum())
```

```python
weight_log_info = weight_log_info.drop(columns=['fat'])
print("Fat column dropped. Remaining columns:", weight_log_info.columns.tolist())
```

### Remove Duplicates

```python
print("Before Daily Activity", len(daily_activity))
daily_activity = daily_activity.drop_duplicates()
print("After Daily Activity", len(daily_activity))

print("Before - Weight Log:", len(weight_log_info))
weight_log_info = weight_log_info.drop_duplicates()
print("After  - Weight Log:", len(weight_log_info))

print("Before - Minute Sleep:", len(sleep_minutes))
sleep_minutes = sleep_minutes.drop_duplicates()
print("After  - Minute Sleep:", len(sleep_minutes))
```

### Remove illogical values

```python
#daily activities where total steps are 0 removed
print("Rows with 0 steps:", (daily_activity['totalsteps'] == 0).sum())
daily_activity = daily_activity[daily_activity['totalsteps'] > 0]
print("Rows remaining after removing 0-step days:", len(daily_activity))
```

```python
#removing any negative values
numeric_cols = ['totalsteps', 'totaldistance', 'calories',
                'veryactiveminutes', 'fairlyactiveminutes',
                'lightlyactiveminutes', 'sedentaryminutes']

for col in numeric_cols:
    neg_count = (daily_activity[col] < 0).sum()
    print(f"Negative values in {col}: {neg_count}")
```

```python
print("Weight range (kg):", weight_log_info['weightkg'].min(), "to", weight_log_info['weightkg'].max())
print("BMI range:", weight_log_info['bmi'].min(), "to", weight_log_info['bmi'].max())
```

### Aggregate Minute sleep to daily level

```python
daily_sleep = sleep_minutes.groupby(['id', 'date_only']).agg(
    total_minutes_asleep = ('value', lambda x: (x == 1).sum()),  # value == 1 means asleep
    total_minutes_in_bed = ('value', 'count')                    # all minutes recorded = in bed
).reset_index()
```

```python
daily_sleep = daily_sleep.rename(columns={'date_only': 'date'})
daily_sleep['date'] = pd.to_datetime(daily_sleep['date'])
```

```python
print(daily_sleep.head())
print(f"Shape: {daily_sleep.shape}")
```

### Merge datasets

```python
#Merge daily activity with daily sleep on Id and Date
activity_sleep = pd.merge(
    daily_activity,
    daily_sleep,
    left_on  = ['id', 'activitydate'],
    right_on = ['id', 'date'],
    how = 'left'
)

# Drop the duplicate date column from sleep
activity_sleep = activity_sleep.drop(columns=['date'])

print(f"Merged dataset shape: {activity_sleep.shape}")
print(activity_sleep.head())
```

```python
activity_sleep['activitydate'] = pd.to_datetime(activity_sleep['activitydate'], format='%d-%m-%Y').dt.strftime('%d-%m-%Y')
weight_log_info['date'] = weight_log_info['date'].dt.strftime('%d-%m-%Y')
```
