```python
#Fixing dates to date time format

# Reload dataframes to ensure original date string formats for conversion
daily_activity = pd.read_csv('/content/dailyActivity_merged.csv')
sleep_minutes = pd.read_csv('/content/minuteSleep_merged.csv')
weight_log_info = pd.read_csv('/content/weightLogInfo_merged.csv')

daily_activity['ActivityDate'] = pd.to_datetime(daily_activity['ActivityDate'], format='%m/%d/%Y')
sleep_minutes['date'] = pd.to_datetime(sleep_minutes['date'], format='%m/%d/%Y %I:%M:%S %p')
weight_log_info['Date'] = pd.to_datetime(weight_log_info['Date'], format='%m/%d/%Y %I:%M:%S %p')

#organising dates from lowest to highest
print("Daily Activity date range:", daily_activity['ActivityDate'].min(), "to", daily_activity['ActivityDate'].max())
print("Sleep Minutes date range:", sleep_minutes['date'].min(), "to", sleep_minutes['date'].max())
print("Weight Log Info date range:", weight_log_info['Date'].min(), "to", weight_log_info['Date'].max())
```

```python
#changing the format of the dates to dd-mm-yyyy
daily_activity['ActivityDate'] = daily_activity['ActivityDate'].dt.strftime('%d-%m-%Y')
sleep_minutes['date'] = sleep_minutes['date'].dt.strftime('%d-%m-%Y')
weight_log_info['Date'] = weight_log_info['Date'].dt.strftime('%d-%m-%Y')
```

### Process Stage

```python
#standardise column names
daily_activity.columns = daily_activity.columns.str.lower().str.replace(' ', '_')
sleep_minutes.columns = sleep_minutes.columns.str.lower().str.replace(' ', '_')
weight_log_info.columns = weight_log_info.columns.str.lower().str.replace(' ', '_')
#verify
print(daily_activity.columns.tolist())
print(sleep_minutes.columns.tolist())
print(weight_log_info.columns.tolist())
```

```python
#Fix data types
#Daily activity - convert date and ensure numeric columns are correct
daily_activity['activitydate']= pd.to_datetime(daily_activity["activitydate"])
daily_activity['day_of_week'] = daily_activity['activitydate'].dt.day_name()
daily_activity['month'] = daily_activity['activitydate'].dt.month_name()

print(daily_activity.dtypes)
```

```python
#weight log
weight_log_info['date']= pd.to_datetime(weight_log_info['date'], format='%d-%m-%Y')
weight_log_info['ismanualreport'] = weight_log_info['ismanualreport'].astype(bool)
print(weight_log_info.dtypes)
```

```python
#minutes sleep
sleep_minutes['date']=pd.to_datetime(sleep_minutes['date'])
sleep_minutes['date_opnly'] = sleep_minutes['date'].dt.date

print(sleep_minutes.dtypes)
```

## Handle Null values

```python
print("Nulls in Daily Activity:")
print(daily_activity.isnull().sum())

print("\nNulls in Sleep Minutes:")
print(sleep_minutes.isnull().sum())

print("\nNulls in Weight Log Info:")
print(weight_log_info.isnull().sum())
```
